# Analysis

Capomulin is a viable drug regiment to reduce tumor growth for the following reasons:

1.) The drug had the most number of mice complete the study with the exception of Remicane.
2.) The data shows a strong correlation between mouse weight and tumor volume, suggesting that mouse weight is a factor in the effectiveness of any drug therapy.
3.) There was one potential outlier in the data within the mice of the Infubinol regimen. (Marked in the box and whisker plot.

```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Setup

study_path = "data/Study_results.csv"
mouse_path = "data/Mouse_metadata.csv"
df_study = pd.read_csv(study_path)
df_mouse = pd.read_csv(mouse_path)

df_main = pd.merge(df_study, df_mouse, how="left", on="Mouse ID")
```

```python
mouse_id_dup = df_main.loc[df_main["Mouse ID"] == "g989"]
mouse_id_dup
```

```python
df_clean = df_main[df_main['Mouse ID'].isin(mouse_id_dup)==False]
df_clean
```

```python
means = df_clean.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
medians = df_clean.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variances = df_clean.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
sds = df_clean.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sems = df_clean.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

sum_table = pd.DataFrame({"Mean Tumor Volume":means,
                              "Median Tumor Volume":medians,
                              "Tumor Volume Variance":variances,
                              "Tumor Volume Std. Dev.":sds,
                              "Tumor Volume Std. Err.":sems})
sum_table
```

```python
sum_table = df_clean.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
sum_table
```

```python
# Quartiles, Outliers, Boxplots

# Final Data Point
max_tumor = df_clean.groupby(["Mouse ID"]).max()
max_tumor = max_tumor.reset_index()
max_tumor
```

```python
df_merge = max_tumor[['Mouse ID','Timepoint']].merge(df_clean,on=['Mouse ID','Timepoint'],how="left")
```

```python
# Outliers of top performing drugs

capomulin = df_merge.loc[df_merge["Drug Regimen"] == "Capomulin"]['Tumor Volume (mm3)']
ramicane = df_merge.loc[df_merge["Drug Regimen"] == "Ramicane"]['Tumor Volume (mm3)']
infubinol = df_merge.loc[df_merge["Drug Regimen"] == "Infubinol"]['Tumor Volume (mm3)']
ceftamin = df_merge.loc[df_merge["Drug Regimen"] == "Ceftamin"]['Tumor Volume (mm3)']

# Capomulin
cap_q = capomulin.quantile([.25,.5,.75])
cap_lq = cap_q[0.25]
cap_uq = cap_q[0.75]
cap_iqr = cap_uq-cap_lq
cap_lb = cap_lq - (1.5*cap_iqr)
cap_ub = cap_uq + (1.5*cap_iqr)
print(f"Capomulin potential outliers: {capomulin.loc[(capomulin < cap_lb) | (capomulin > cap_ub)]}")

# Ramicane
ram_q = ramicane.quantile([.25,.5,.75])
ram_lq = ram_q[0.25]
ram_uq = ram_q[0.75]
ram_iqr = ram_uq-ram_lq
ram_lb = ram_lq - (1.5*ram_iqr)
ram_ub = ram_uq + (1.5*ram_iqr)
print(f"Ramicane potential outliers: {ramicane.loc[(ramicane < ram_lb) | (ramicane > ram_ub)]}")

# Infubinol
inf_q = infubinol.quantile([.25,.5,.75])
inf_lq = inf_q[0.25]
inf_uq = inf_q[0.75]
inf_iqr = inf_uq-inf_lq
inf_lb = inf_lq - (1.5*inf_iqr)
inf_ub = inf_uq + (1.5*inf_iqr)
print(f"Infubinol potential outliers: {infubinol.loc[(infubinol < inf_lb) | (infubinol > inf_ub)]}")

# Ceftamin
cef_q = ceftamin.quantile([.25,.5,.75])
cef_lq = cef_q[0.25]
cef_uq = cef_q[0.75]
cef_iqr = cef_uq-cef_lq
cef_lb = cef_lq - (1.5*cef_iqr)
cef_ub = cef_uq + (1.5*cef_iqr)
print(f"Ceftamin potential outliers: {ceftamin.loc[(ceftamin < cef_lb) | (ceftamin > cef_ub)]}")
```

```python
# Box Plot - infubinal outlier marked

out = dict(markerfacecolor='red',markersize=12)
plt.boxplot([capomulin,ramicane,infubinol,ceftamin],labels=['Capomulin','Ramicane','Infubinol','Ceftamin'],flierprops=out)
plt.ylabel('Final Tumor Volume (mm3)')

plt.show()
```

```python
# Line Plot Capomulin

cap_table = df_clean.loc[df_clean['Drug Regimen'] == "Capomulin"]
mousedata = cap_table.loc[cap_table['Mouse ID']== 'l509']

plt.plot(mousedata['Timepoint'],mousedata['Tumor Volume (mm3)'])

plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse l509')

plt.show()

```

```python
# Scatter Plot Capomulin

cap_average = cap_table.groupby(['Mouse ID']).mean()

plt.scatter(cap_average['Weight (g)'], cap_average['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```

```python
# Correlation and Regression - Capomulin

corr=round(st.pearsonr(cap_average['Weight (g)'], cap_average['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and the average tumor volume is {corr}")

model = st.linregress(cap_average['Weight (g)'], cap_average['Tumor Volume (mm3)'])
y_values = cap_average['Weight (g)']*model[0]+model[1]
plt.scatter(cap_average['Weight (g)'],cap_average['Tumor Volume (mm3)'])
plt.plot(cap_average['Weight (g)'],y_values,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```
