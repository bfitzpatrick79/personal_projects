```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# you and a friend make a bet about climbing a building. Starting at step 0, you will roll two dice.
# if the combo rolls 2-5, you stay where you are. If you roll 6-9, you go up one step. If you roll 10-12, you go down one step
# unless you are at step 0. Will you reach the top of a building with N stairs?

#extending the problem, there is a 0.01% chance of slipping and rolling down the stairs and returning to 0.
```

```python
end_steps = []
for i in range(15) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7) + np.random.randint(1,7)
        if dice <= 5:
            step = step
        elif dice <= 9:
            step = step + 1
        else:
            step = max(0, step - 1)
       
        if np.random.rand() <= 0.010 :
            step = 0
        random_walk.append(step)
    end_steps.append(random_walk)
```

```python
np_aw_t = np.transpose(np.array(end_steps))
plt.plot(np_aw_t)
plt.show
```

```python
ends = np_aw_t[-1,:]
```

```python
plt.hist(ends, bins=100)
plt.show()
```
