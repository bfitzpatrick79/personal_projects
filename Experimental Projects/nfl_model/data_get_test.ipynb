```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
team = 'chi'
year_range = range(2000,2020)
bear_dict = {}
```

```python
# produces dictionary of dataframes for years in year_range, then adds a column specifying year
for year in year_range:
    print(year)
    url = f'https://www.pro-football-reference.com/teams/{team}/{year}.htm'
    bear_data = pd.read_html(url, header=1)
    bear_dict[year] = bear_data[1]
    bear_dict[year]['year'] = year
```

```python
bear_dict[2000].head(3)
```

```python
# appends all separate dataframes into a single frame
full_bear = pd.DataFrame()
for i in bear_dict:
    full_bear = full_bear.append(bear_dict[i], ignore_index=False)
```

```python
full_bear.head(10)
```

```python
for i in full_bear.columns:
    print(i)
```

```python
# Dataframe Cleanup
column_clean_dict =  {'Unnamed: 3':'Kickoff_time',
     'Unnamed: 5': 'W/L',
     'Unnamed: 8' : 'Home/Away',
     'Tm' : 'Team_Score',
     'Opp.1' : 'Oppo_score',
     '1stD' : '1D_gained',
     'TotYd' : 'Team_Total_Yards',
     'PassY' : 'Team_Pass_Yards',
     'RushY' : 'Team_Rush_Yards',
     'TO' : 'team_turnover',
     '1stD.1' : '1D_allowed',
     'TotYd.1' : 'oppo_total_yards',
     'PassY.1' : 'oppo_pass_yards',
     'RushY.1' : 'oppo_rush_yards',
     'TO.1' : 'oppo_turnover',
     'Offense' : 'offense_expected_points',
     'Defense' : 'defense_expected_points',
     'Sp. Tms' : 'sp_expected_points'
    }
```

```python
# Mark bye weeks before filling NaNs in other columns.

full_bear[full_bear['Opp']=='Bye Week'].fillna(value='BYE', inplace=True, axis=0)
```

```python
full_bear.rename(column_clean_dict, inplace=True, axis=1)

full_bear.drop('Unnamed: 4', axis=1, inplace=True)

full_bear['OT'].fillna(value=0, inplace=True)
full_bear['OT'].replace({'OT': 1}, inplace=True)

full_bear['Home/Away'].fillna(value='HOME', inplace=True)
full_bear['Home/Away'].replace({'@': 'AWAY'}, inplace=True)

full_bear['team_turnover'].fillna(value=0, inplace=True)
full_bear['oppo_turnover'].fillna(value=0, inplace=True)
full_bear.head(10)
```
