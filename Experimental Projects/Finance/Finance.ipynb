```python
import datetime as dt
```

```python
import matplotlib.pyplot as plt
import mplfinance as mpl
import matplotlib.dates as mdates
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12.0, 12.0)
```

```python
import pandas as pd
import pandas_datareader.data as web
```

```python
start = dt.datetime(2000,1,1)
```

```python
end = pd.to_datetime("today")
print(end)
```

```python
data = web.DataReader(['TSLA'], 'yahoo', start, end)
data.head()
```

```python
data[['Close', 'High', 'Low', 'Open']].plot(figsize=(15, 15), linewidth=0.5, title='Apple Prices')
```

```python
data['100ma'] = data['Adj Close'].rolling(window=100).mean()
```

```python
data['100ma'].plot(figsize = (12,12))
```

```python
ax1 = plt.subplot2grid((6, 1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6, 1), (5,0), rowspan=1, colspan=1, sharex=ax1)
ax1.plot(data.index, data['Adj Close'])
ax1.plot(data.index, data['100ma'])
ax2.plot(data.index, data['Volume'])
```

```python
data_ohlc = data['Adj Close'].resample('W').ohlc()
data_vol = data['Volume'].resample('W').sum()
```

```python
data_ohlc.plot(linewidth=0.5)
data_vol.plot(linewidth=0.5)
```
