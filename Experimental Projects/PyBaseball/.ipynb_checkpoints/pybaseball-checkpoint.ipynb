```python
cubs_df = schedule_and_record(2016, 'CHC')
```

```python
cubs_df.tail()
```

```python
cubs_df.shape
```

```python
cubs_df['R'].plot()
plt.title('2016 runs per game')
plt.axhline(cubs_df['R'].mean(), linewidth=2, linestyle='--', color='r')
plt.show()
```

```python
cubs_df['Attendance'].plot()
plt.axhline(cubs_df['Attendance'].mean(), linewidth=2, linestyle='--', color='r')
plt.title('Attendence')
plt.show()
```

```python
cubs_df['Streak'].plot()
plt.axhline(cubs_df['Streak'].mean(), linewidth=2, linestyle='--', color='r')
plt.title('Streak over Time')
```

```python
cubs_df['W/L'].value_counts().plot(kind='bar', color=['b', 'r', 'grey', 'grey', 'grey'])
```
