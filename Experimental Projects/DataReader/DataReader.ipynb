```python
from pandas_datareader.data import DataReader
from datetime import date
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Set start and end dates
start = date(2008, 1, 1)
stop = date(2011,12,31)

# Set the ticker
series = ['NASDAQCOM', 'NASDAQ100']

# Set the data source
data_source = 'fred'

# Import the stock prices
stock_markets = DataReader(series, data_source, start, stop)

# Display and inspect the result
print(stock_markets.head())


stock_markets.plot(subplots=True)
plt.xticks(rotation=90)
plt.show()
```
